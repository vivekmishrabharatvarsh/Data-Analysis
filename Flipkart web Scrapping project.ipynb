```python
# navigating to next page
import pandas as pd
product_name=[]
prices=[]
review=[]
memory=[]
Screen=[]
Camera=[]
Battery=[]
Processor=[]
for i in range(1,9):
    u='https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&p%5B%5D=facets.brand%255B%255D%3Drealme&p%5B%5D=facets.availability%255B%255D%3DExclude%2BOut%2Bof%2BStock&param=7564&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIlNob3AgTm93Il0sInZhbHVlVHlwZSI6Ik1VTFRJX1ZBTFVFRCJ9fX19fQ%3D%3D&otracker=clp_metro_expandable_1_5.metroExpandable.METRO_EXPANDABLE_Shop%2BNow_mobile-phones-store_Q1PDG4YW86MF_wp2&fm=neo%2Fmerchandising&iid=M_0fd2c8fe-0a36-487a-8680-f334c6f04082_5.Q1PDG4YW86MF&ppt=browse&ppn=browse&ssid=aia9y71d9s0000001684238360087&page='+str(i)
    website1= requests.get(u,headers=headers)
    sou= BeautifulSoup(website1.content,"html.parser")
    soup1=BeautifulSoup(sou.prettify(),"html.parser")
    np=soup1.find("a",class_='_1LKTO3').get('href')
    link='https://www.flipkart.com'+np
    
    #Finding Product name for all items
    
    title=soup1.find_all("div",class_='_4rR01T')

    for i in title:
        name=i.text
        product_name.append(name.strip())

    #Finding Prices for all items
  
    pr=soup1.find_all(class_='_30jeq3 _1_WHN1')
    for i in pr:
        name=i.text
        prices.append(name.strip()[1:])

    #Finding Reviews for all items

    box=soup1.find(class_='_1YokD2 _3Mn1Gg')
    re=box.find_all(class_='_3LWZlK')
    for i in re:
        name=i.text
        review.append(name.strip())
    #Finding details for all items
    l=[]
    me=soup1.find_all(class_='rgWa7D')
    for i in me:
        name=i.text
        l.append(name.strip())
    #Memory
    memory=l[0]
    #Screen
    Screen=l[1]
    #Camera
    Camera=l[2]
    #Battery
    Battery=l[3]
    #Processor
    Processor=l[4]

df=pd.DataFrame({"Product Name":product_name,"prices":prices,"review":review,"memory":memory,"Screen":Screen,"Camera":Camera,
                "Battery":Battery,"Processor":Processor})

df.to_csv('E:\Flipkart web Scrapping project.csv')
    
```
